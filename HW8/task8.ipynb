```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def get_closest_variance(var_i, varRetained):
    var_i_cmp = var_i[var_i >= varRetained]
    if var_i_cmp.size > 0:
        return 

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * F.dot(np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    #U, d, V = посчитать SVD матрицы Sigma
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)  # trace
    var_i = np.array([np.sum(d[: i + 1]) / dTot * 100.0 for i in range(n)])

    #m = вычислите необходимое число главных компонент
    m = np.argmax(var_i >= varRetained * 100) + 1
    print(m)
    print '{:.2f} variance retained in {} dimensions'.format(var_i[m - 1], m)
    print(var_i[m - 1:m+2])
    #% верните число m и точность, которая достигается при этом числе главных компонент

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    #U_reduced = только m главных компонент
    U_reduced = U[:, :m]
    #G = вычислить матрицу в преобразованном пространстве
    G = U_reduced.dot(np.diag(d)[:m, :m])
    
    return G, U_reduced
```

```python
best_adjusted_mutual = 0
best_eps = 0
best_min_sampler = 0
for eps_par in np.arange(4, 11, 0.2):
    for min_samples_par in np.arange(1, 20, 1):
        db = DBSCAN(eps=eps_par, min_samples=min_samples_par).fit(X)
        core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
        core_samples_mask[db.core_sample_indices_] = True
        labels = db.labels_
        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        cur_adj = adjusted_rand_score(labels, y)
        if cur_adj > best_adjusted_mutual and n_clusters_ == 10:
            best_eps = eps_par
            best_min_sampler = min_samples_par
            best_adjusted_mutual = cur_adj
        print "eps =", eps_par, "min_samples_par =", min_samples_par
        print "    cur_adj", cur_adj, "n_clsters", n_clusters_, "noize", np.sum(labels == -1)
```

## KMeans

```python
def render_image(num_of_colors, image_array, shape_tuple):
    kmeans = KMeans(n_clusters=num_of_colors, random_state=0).fit(image_array)
    kmeans.cluster_centers_
    new_image_array = np.array(list(map(lambda lab: kmeans.cluster_centers_[lab], kmeans.predict(image_array))))
    show_image(new_image_array.reshape(shape_tuple))
    return new_image_array
```

```python
kmeans_image_array = []
for col in color_array:
    kmeans_image_array.append(render_image(col, image_array, (w, h, d)))
```

## AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
```

```python
def render_image_agg(num_of_colors, image_array, shape_tuple):
    knn_graph = kneighbors_graph(image_array, 10, include_self=False)
    aggl = AgglomerativeClustering(n_clusters=num_of_colors, connectivity=knn_graph)
    a = aggl.fit_predict(image_array)
    cluster_centers_ = np.zeros((np.unique(a).size, 3))
    for i in np.unique(a):
        cluster_centers_[i] = image_array[a == i].mean(axis=0)
        
    new_image_array = np.array(list(map(lambda lab: cluster_centers_[lab], a)))
    show_image(new_image_array.reshape(shape_tuple))
    return new_image_array
```

```python
aggl_images = []
for col in color_array:
    aggl_images.append(render_image_agg(col, image_array, (w, h, d)))
```
