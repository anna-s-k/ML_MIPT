```python
import numpy as np
import scipy.stats as sps
from sklearn import metrics

import matplotlib.pyplot as plt
from scipy.stats import uniform, norm

from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae

from scipy.optimize import minimize
import pandas as pd

%matplotlib inline
```

## Генерация данных

```python
dataset = {}
```

```python
dataset['data'] = uniform.rvs(size = 500, loc=-3, scale=6, random_state=2)
dataset['target'] = 0.5 * dataset['data'] + 1 + norm.rvs(size=500, loc=0, scale=0.2, random_state=2)
```

```python
plt.figure(figsize=(16, 7))
plt.scatter(dataset['data'], dataset['target'], s=20)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## Минимизируем MSE

```python
def MSE(param):
    return mse(dataset['target'], param[1] * dataset['data'] + param[0])
```

```python
minimization_result = minimize(MSE, [1, 1])
```

```python
print (minimization_result)
```

Есл

```python
plt.figure(figsize=(16, 7))
plt.scatter(dataset['data'], dataset['target'], s=20)
target_pred = minimization_result.x[1] * dataset['data'] + minimization_result.x[0]
plt.plot(np.sort(dataset['data']), np.sort(target_pred), c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## Теперь добавим в выборку 75 точек -1 + $\varepsilon$

```python
dataset_spoiled = {}
```

```python
dataset_spoiled['data'] = np.append(dataset['data'], uniform.rvs(size=75, loc=-3, scale=6, random_state=2))
dataset_spoiled['target'] = np.append(dataset['target'], -1 * np.ones(75) + norm.rvs(size=75, 
                                                                                     scale=0.2, random_state=2))
```

#### MSE

```python
def MSE_sp(param):
    return mse(dataset_spoiled['target'], param[1] * dataset_spoiled['data'] + param[0])
```

```python
minimization_result_mse = minimize(MSE_sp, [1, 1])
```

```python
print (minimization_result_mse)
```

#### MAE

```python
def MAE_sp(param):
    return mae(dataset_spoiled['target'], param[1] * dataset_spoiled['data'] + param[0])
```

```python
minimization_result_mae = minimize(MAE_sp, [1, 1])
```

```python
print (minimization_result_mae)
```

```python
plt.figure(figsize=(16, 7))
plt.scatter(dataset_spoiled['data'], dataset_spoiled['target'], s=20)
target_pred_mse = minimization_result_mse.x[1] * dataset_spoiled['data'] + minimization_result_mse.x[0]
target_pred_mae = minimization_result_mae.x[1] * dataset_spoiled['data'] + minimization_result_mae.x[0]
plt.plot(np.sort(dataset_spoiled['data']), np.sort(target_pred_mse), c='r', label='MSE')
plt.plot(np.sort(dataset_spoiled['data']), np.sort(target_pred_mae), c='g', label='MAE')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
pd.DataFrame({"Реальные коэффициенты": [1, 0.5],
            "MSE на выборке без выбросов": minimization_result.x, 
              "MSE на выборке с выбросами": minimization_result_mse.x,
             "MAE на выборке с выбросами": minimization_result_mae.x})
```

Видим, что MAE устойчивее к выборосам, т.к. полученные коэффициенты очень близки к реальным
